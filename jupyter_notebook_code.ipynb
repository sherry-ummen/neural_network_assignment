```
import sys
import pandas as pd
import numpy as np

# We want to use MLPRegressor from sklearn
from sklearn.neural_network import MLPRegressor

datafile = 'users_elisa.csv'

data = pd.read_csv(datafile, header=0, delimiter=';', decimal=',')
trainy = data.iloc[:, 5:12].values
trainx = data.iloc[:, 13:55].values

product_model = MLPRegressor(hidden_layer_sizes=(5,),
                              activation='relu',
                              solver='adam',
                              learning_rate='adaptive',
                              max_iter=1500,
                              learning_rate_init=0.001,
                              alpha=0.01, verbose=True)
product_model.fit(trainx, trainy)
print(product_model.score(trainx, trainy))
ypred = product_model.predict(trainx)
print(ypred)  # print to a csv file
```

```
import pandas as pd
import numpy as np
import sys

# We want to use MLPRegressor from sklearn
from sklearn.neural_network import MLPRegressor

datafile = 'users_elisa.csv'


data = pd.read_csv(datafile, header=0, delimiter=';', decimal=',')

trainy = data.iloc[:, 1:3].join(data.iloc[:, 5:12])
trainx = data.iloc[:, 13:55].values

product_model = MLPRegressor(hidden_layer_sizes=(5,),
                              activation='relu',
                              solver='adam',
                              learning_rate='adaptive',
                              max_iter=1500,
                              learning_rate_init=0.01,
                              alpha=0.01, verbose=True)
product_model.fit(trainx, trainy)
print(product_model.score(trainx, trainy))
ypred = product_model.predict(trainx)
print(ypred)  # print to a csv file
```

```
import pandas as pd
import numpy as np
import sys

# We want to use MLPRegressor from sklearn
from sklearn.neural_network import MLPRegressor

datafile = 'users_elisa.csv'

np.random.seed(100)
data = pd.read_csv(datafile, header=0, delimiter=';', decimal=',')

trainy = data.iloc[:, 4].values
trainx = data.iloc[:, 1:3].join(data.iloc[:, 5:55]).values

product_model = MLPRegressor(hidden_layer_sizes=(5,),
                              activation='relu',
                              solver='adam',
                              learning_rate='adaptive',
                              max_iter=1500,
                              learning_rate_init=0.001,
                              alpha=0.01, verbose=False)
product_model.fit(trainx, trainy)
print(product_model.score(trainx, trainy))
ypred = product_model.predict(trainx)
print(ypred)  # print to a csv file
pd.DataFrame(product_model.loss_curve_).plot()
```
